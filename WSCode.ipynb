```python
browser = Browser('chrome')
driver = webdriver.Chrome()
url2 = 'https://www.basketball-reference.com/leaders/ws_season.html'
driver.get(url2)
browser.visit(url2)
html = browser.html
wait = WebDriverWait(driver, 20) 
soup = BeautifulSoup(html, 'html.parser')
```

```python
div = soup.find('div', class_ = 'table_container is_setup')
table = div.find('table')
body = table.find('tbody')
TR = body.find_all('tr')
```

```python
players = []
WinShares = []
year = []
for t in TR:
    info = (t.text)
    info2 = info.split('\n')
    names = info2[1].replace('*','')
    WS = info2[2]
    Year = info2[3]
    players.append(names)
    WinShares.append(WS)
    year.append(Year)
```

```python
div = soup.find('div', class_ = 'table_container is_setup')
table = div.find('table')
body = table.find('tbody')
TR = body.find_all('tr')
TR
```

```python
position1 = []
for t in range(1,10):
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        info = (TR[t])
        info2 = (info.text)
        info3 = info2.split('\n')
        name = info3[1].replace('*','')
        link_element = driver.find_element(By.LINK_TEXT, name)
        driver.execute_script("arguments[0].click()", link_element)
        new_page_html = driver.page_source
        soup = BeautifulSoup(new_page_html, 'html.parser')
        positions = soup.find_all('td', class_ = 'center')
        finalposition = positions[1].text
        position1.append(finalposition)
        driver.back()
        previous_page_html = driver.page_source
        soup = BeautifulSoup(previous_page_html, 'html.parser')
    except:
        continue 
```
