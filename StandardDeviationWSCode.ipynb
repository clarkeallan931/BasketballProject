```python
url1 = 'https://www.basketball-reference.com/teams/MIL/1971.html'
url2 = 'https://www.basketball-reference.com/teams/BOS/1986.html'
url3 = 'https://www.basketball-reference.com/teams/LAL/2000.html'
url4 = 'https://www.basketball-reference.com/teams/MIA/2013.html'
url5 = 'https://www.basketball-reference.com/teams/HOU/2018.html'
```

```python
browser = Browser('chrome')
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)
driver.get(url1)
browser.visit(url1)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
di = soup.find('div', id = 'div_advanced')
TR = di.find_all('td', class_ = 'right')
```

```python
data=[]
for t in TR:
    data.append(t.text)
```

```python
data
```

```python
Bucks1971Winshares = []
for h in range(17,len(data),24):
    Bucks1971Winshares.append(data[h])
```
